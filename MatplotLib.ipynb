## Simple Plotting

If you only provide one array of values, MatplotLib assumes they're a sequence of values starting from 0:

```python
import numpy as np
import matplotlib.pyplot as plt

my_array = np.array([1, 2, 5])
plt.plot(my_array)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

But you can also pass two arrays for X and Y values.

Also, the third argument is the color + type of plotting. For example, "ro" stands for red circles, while "b-" stands for blue line:

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter plot
X = np.array([1, 2, 3])
Y = np.array([2, 5, 20])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X, Y, 'ro')
plt.show()

# Line plot
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X, Y, 'b-')
plt.show()

# Mixing 3 different values
X1 = np.array([1, 1.5, 3])
Y1 = np.array([2, 8, 20])
X2 = np.array([1, 2.5, 4])
Y2 = np.array([2, 4.4, 28])
X3 = np.array([1, 2, 3])
Y3 = np.array([5, 5, 5])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X1, Y1, 'r^', X2, Y2, 'b-', X3, Y3, 'gs')
plt.show()
```

## Scatter Plot

The `scatter` function also draws scattered plots:

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3])
Y = np.array([2, 3, 4])
figure = plt.scatter(X, Y)
plt.xlabel('x axis')
plt.ylabel('y axis')

# Set properties to the figure
plt.setp(figure, color='r')

plt.show()
```

## Subplots

Matplotlib allows you to pass categorical variables directly to many plotting functions:

```python
import matplotlib.pyplot as plt

names = ['Jan', 'Feb', 'Mar']
values = [56, 83, 136]

plt.figure(1, figsize=(9, 3))

# Position 1 of a larger plot with 1 row and 3 columns
plt.subplot(131)
plt.bar(names, values)

# Position 2 of a larger plot with 1 row and 3 columns
plt.subplot(132)
plt.scatter(names, values)

# Position 3 of a larger plot with 1 row and 3 columns
plt.subplot(133)
plt.plot(names, values, linewidth=3.0)

plt.suptitle('Precipitação média em Maceió')
plt.show()
```

Or you can use `sublplot` with the values as different arguments and set properties on each figure:

```python
import matplotlib.pyplot as plt

names = ['Jan', 'Feb', 'Mar']
values = [56, 83, 136]

plt.figure(1, figsize=(7, 7))

# Position 1 of a larger plot with 2 row and 2 columns
fig1 = plt.subplot(2, 2, 1)
fig1.set_title("Bars")
plt.bar(names, values)

# Position 2 of a larger plot with 2 row and 2 columns
fig2 = plt.subplot(2, 2, 2)
fig2.set_title("Circles")
plt.scatter(names, values)

# Position 3 of a larger plot with 2 row and 2 columns
fig3 = plt.subplot(2, 2, 3)
fig3.set_title("Line")
plt.plot(names, values)

plt.suptitle('Precipitação média em Maceió')
plt.show()
```
