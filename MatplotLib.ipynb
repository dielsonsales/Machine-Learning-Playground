If you only provide one array of values, MatplotLib assumes they're a sequence of values starting from 0:

```python
import numpy as np
import matplotlib.pyplot as plt

my_array = np.array([1, 2, 5])
plt.plot(my_array)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

But you can also pass two arrays for X and Y values.

Also, the third argument is the color + type of plotting. For example, "ro" stands for red circles, while "b-" stands for blue line:

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3])
Y = np.array([2, 5, 20])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X, Y, 'ro')
plt.show()

plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X, Y, 'b-')
plt.show()
```

It's also possible to combine different arrays in the same pl

```python
import numpy as np
import matplotlib.pyplot as plt

X1 = np.array([1, 2, 3])
Y1 = np.array([2, 5, 20])
X2 = np.array([1, 2.5, 4])
Y2 = np.array([2, 4.4, 28])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(X1, Y1, 'ro', X2, Y2, 'b-')
plt.show()
```

The `scatter` function also draws scattered plots:

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3])
Y = np.array([2, 3, 4])
plt.scatter(X, Y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
